## Answer 2-a
### How many orders were shipped by Speedy Express in total?

## Answer 2-b.
### What is the last name of the employee with the most orders?

`SELECT e.LastName FROM Orders o, Employees e where o.EmployeeID=e.EmployeeID group by o.EmployeeID order by count(o.EmployeeID) desc limit 1;`

LastName is Peacock

## Answer 2-c
### What product was ordered the most by customers in Germany?

`SELECT p.ProductName FROM Orders o, OrderDetails d, Customers c, Products p where o.OrderID=d.OrderID and o.CustomerID=c.CustomerID and p.ProductID=d.ProductID and c.Country='Germany'group by d.ProductID order by sum(Quantity) desc limit 1;`

Boston Crab Meat was most ordered by customers in Germany.
